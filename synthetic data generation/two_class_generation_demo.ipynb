# This notebooks demonstrates three types of synthetic data in a simple-to-complex order.

```python
from two_class_data_generation import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(data, style='.'):
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.plot(data[i], style)
        if i<2:
            plt.title("Class 0")
        else:
            plt.title("Class 1")
        plt.ylim(data.min()-.3, data.max()+.3)
    plt.tight_layout()    
```

# 1 Binary
## 1.1 Simplest case: same sigma

```python
data, class_labels = generate_two_class_data(4, 128, binary=True, same_sigma=True)
```

```python
plot(data)
```

```python
class_labels
```

## 1.2 A bit more difficult: different sigmas

```python
data, class_labels = generate_two_class_data(4, 128, binary=True, same_sigma=False)
```

```python
plot(data)
```
